```python
# Install the Python JSONPath library
# %pip install jsonpath-ng

import json
from jsonpath_ng import parse

with open("kobo_output/record_136_raw.json", "r", encoding="utf-8") as f:
    record = json.load(f)

print(record.keys())
```

```python
#vCreate a helper function:
from jsonpath_ng import parse

def jp(data, path, first=False, default=None):
    matches = [match.value for match in parse(path).find(data)]
    if first:
        return matches[0] if matches else default
    return matches
```

```python
# testing fields to extract
title = jp(record, "$['dublin_core_group/title']", first=True)
dataset_id = jp(record, "$['control_group/id']", first=True)
description = jp(record, "$['dublin_core_group/description']", first=True)

print(dataset_id)
print(title)
print(description)
```
